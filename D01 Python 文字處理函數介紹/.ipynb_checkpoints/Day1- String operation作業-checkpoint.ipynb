## 讀入所需文件

```python
import pandas as pd
#TSV文件有別於CSV的文件是使用\t作为分隔符，CSV則是使用,作为分隔符。
dataset = pd.read_csv(r'Restaurant_Reviews.tsv', sep='\t')
dataset.head(3)
```

```python
all_review = dataset['Review'].values
```

---

### 計算有多少個句子是以 . 結尾

```python
n=0

for sentence in all_review:
    if sentence.endswith('.'):
        n += 1

print('There are {} sentences end with .'.format(n))
```

### 將所有. 換成 ,

```python
all_review = [sentence.replace('.', ',') for sentence in all_review]
```

```python
all_review
```

### 將所有sentence 中的 the 置換成 The

```python
all_review = [sentence.replace('the', 'The') for sentence in all_review]
```

```python
all_review
```

### 將偶數句子全部轉換為大寫，基數句子全部轉換為小寫

```python
all_review = [sentence.upper() if (enu+1) % 2 == 0 else sentence.lower() for enu, sentence in enumerate(all_review)]
```

```python
all_review
```

### 將所有句子合併在一起，並以' / ' 為間隔

```python
'/'.join(all_review)
```
